```python
import yfinance as yf

def fetch_bitcoin_data():
    # Ticker symbol for Bitcoin is 'BTC-USD'
    ticker = "BTC-USD"

    # Fetch historical data for Bitcoin
    bitcoin_data = yf.download(ticker, start="2022-01-01", end="2022-12-31")

    return bitcoin_data

# Fetch Bitcoin data
bitcoin_data = fetch_bitcoin_data()

# Print the fetched data
print(bitcoin_data)
```

# **To extract data for today with a 5-minute interval**

```python

import yfinance as yf

# Define the ticker symbol and interval
ticker = "BTC-USD"
interval = "5m"

# Fetch the data
today_data = yf.download(ticker, period="1d", interval=interval)

# Print the data
print(today_data)
```
